```python
import snap
import pandas as pd
import numpy as np
```

```python
#Importanto o grafo
G = snap.LoadEdgeList(snap.PNGraph, "grafo.txt", 0, 1)
```

```python
#Page Rank
PRankH = snap.TIntFltH()
snap.GetPageRank(G, PRankH,100,1000)
PRankH.SortByDat(False) #ordenando os dados
for item in PRankH:
    print item, PRankH[item]
```

```python
#Salvando resultados
FOut = snap.TFOut("pagerank_result.bin")
PRankH.Save(FOut)
```
